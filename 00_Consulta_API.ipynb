```python
import requests
import pandas as pd
```

```python
response_films = requests.get('https://ghibliapi.vercel.app/films')
response_films.raise_for_status()

films = response_films.json()
#print(films)

response_people = requests.get('https://ghibliapi.vercel.app/people')
response_people.raise_for_status()

people = response_people.json()
#print(people)

response_species = requests.get('https://ghibliapi.vercel.app/species')
response_species.raise_for_status()

species = response_species.json()
#print(species)
```

```python
df_films = pd.DataFrame.from_dict(films)
df_people = pd.DataFrame.from_dict(people)
df_species = pd.DataFrame.from_dict(species)
df_species


```
